```python
useful_df = df.drop(['new_tests', 'total_tests', 'total_tests_per_thousand',
       'new_tests_per_thousand', 'new_tests_smoothed',
       'new_tests_smoothed_per_thousand', 'tests_per_case', 'positive_rate',
       'tests_units', 'handwashing_facilities'], axis = 1)
```

```python
useful_df.isnull().sum()
```

```python
# replacing nan values of new cases and new death to zero (fillna)
replaced_df = useful_df.fillna({
    
    'new_cases':0,
    'new_cases_smoothed':0,
    'new_deaths':0,
    'new_deaths_smoothed':0,
    'new_cases_per_million':0,
    'new_cases_smoothed_per_million':0,
    'new_deaths_per_million':0,
    'new_deaths_smoothed_per_million':0
    
})
```

```python
replaced_df.isnull().sum()
```

```python
#using forwardfill for total_cases, total_deaths, total_cases_per_million, total_deaths_per_million
replaced_df[['total_cases', 'total_deaths', 'total_cases_per_million', 'total_deaths_per_million']] = replaced_df[['total_cases', 'total_deaths', 'total_cases_per_million', 'total_deaths_per_million']].fillna(method = 'ffill')
```

```python
replaced_df.head(1000)
```

```python
replaced_df.isnull().sum()
```

```python
replaced_df.drop(['hospital_beds_per_thousand', 'extreme_poverty', 'stringency_index', 'female_smokers', 'male_smokers'], axis = 1, inplace=True)
```

```python
pd.set_option('display.max_columns', None)
replaced_df.head(100)
```

```python
#Selecting the locaion to nepal
nepal = replaced_df.loc[replaced_df['location'] == 'Nepal'  ]
```

```python
nepal
```

```python
#Total cases per continent
pv_table = pd.pivot_table(replaced_df, index=replaced_df['date'], values='total_cases', columns='continent')
plt.figure(figsize=(5,10))
sns.heatmap(pv_table, cmap="YlGnBu")
```

```python
plt.figure(figsize=(18,5))
sns.barplot(x = 'continent', y = 'total_cases_per_million', data = replaced_df,)
```

```python
# New cases by data
pv_table = pd.pivot_table(nepal, index=nepal['date'], values='new_cases', columns='location')
plt.figure(figsize=(5,10))
sns.heatmap(pv_table, cmap="YlGnBu")
```

```python
#Total cases by date
pv_table = pd.pivot_table(nepal, index=nepal['date'], values='total_cases', columns='location')
plt.figure(figsize=(5,10))
sns.heatmap(pv_table, cmap="YlGnBu")
```
