# Task 2: From the ‘Iris’ dataset, predicting the optimum number of clusters and representing it visually using K-means clustering (Unsupervised Learning)

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### importing the data

```python
#loading the dataset
df = pd.read_csv('Datasets/Iris.csv')
```

```python
#overview of the dataset
df.head()
```

```python
df.columns
```

## Visual Data Analysis

### Pairplot of the following features

```python
sns.pairplot(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])
```

## Importing KMeans

```python
from sklearn.cluster import KMeans
```

## Elbow Method, for selecting the appropriate 'K' value

```python
distribution = []
K = range(1,10)

for k in K:
    km = KMeans(n_clusters = k)
    km.fit(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])
    distribution.append(km.inertia_)
```

```python
plt.plot(K,distribution)
plt.xlabel('K')
plt.ylabel('distribution')
plt.title('elbow')

plt.show()
```

### Selecting the number of clusters and training the data

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])

y_kmeans = kmeans.fit_predict(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])
```

```python
df['cluster'] = y_kmeans
df.head()
```

## Visual representation of the predictions (Creating separate data frames for different clusters)

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1.SepalLengthCm , df1.SepalWidthCm, color = 'Red')
plt.scatter(df2.SepalLengthCm , df2.SepalWidthCm, color = 'Green')
plt.scatter(df3.SepalLengthCm , df3.SepalWidthCm, color = 'Blue')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color = 'yellow', marker = "o", label='centroid')

plt.legend()
```

## Another way of visually representing of the predictions 

```python
df['k_means']=kmeans.predict(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])

fig, axes = plt.subplots(1, 1, figsize=(6,4))
axes.scatter(df['SepalLengthCm'], df['SepalWidthCm'], c=df['k_means'], cmap = 'rainbow')

axes.set_title('K_means_cluster', fontsize=18)
```
