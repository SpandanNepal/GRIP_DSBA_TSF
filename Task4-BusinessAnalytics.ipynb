## Visual Data Analysis

### Pairplot of the given features

```python
sns.pairplot(useful_df)
```

### Heatmap, of the given features w.r.t correlation_data

```python
correlation_data = useful_df.corr()
```

```python
sns.heatmap(correlation_data, xticklabels= correlation_data.columns, yticklabels= correlation_data.columns)
```

```python
correlation_data
```

### Countplot for different 'region' w.r.t Sub-Category

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Sub-Category', data=useful_df, hue='Region',palette='viridis')
```

### Countplot for different states in the dataset

```python
plt.figure(figsize=(15,5))
sns.countplot(x='State',data=useful_df)
plt.xticks(rotation=90)
```

### Barplot: 'Category' vs 'Profit', estimator: Standard Deveation (np.std)

```python
sns.barplot(x = 'Category', y = 'Profit', data = useful_df, estimator= np.std )
```

### Barplot: 'Category' vs 'Sales', estimator: Mean (np.mean)

```python
sns.barplot(x = 'Category', y = 'Sales', data = useful_df, estimator=np.mean)
```

### Barplot: 'Sub-Category' vs 'Sales', estimator: mean (np.mean)

```python
plt.figure(figsize=(18,5))
sns.barplot(x = 'Sub-Category', y = 'Sales', data = useful_df, estimator=np.mean)
```

### Piechart, Count of different 'Region'

```python
plt.figure(figsize=(10,10))
plt.pie(useful_df['Region'].value_counts(), labels = useful_df['Region'].drop_duplicates(), startangle = 90, autopct='%.1f%%', wedgeprops={'edgecolor':'white'})
plt.show()
```

### Piechart, Count of the different 'Category'

```python
plt.figure(figsize=(10,10))
plt.pie(useful_df['Category'].value_counts(), labels = useful_df['Category'].drop_duplicates(), 
        startangle = 90, autopct='%.1f%%', wedgeprops={'edgecolor':'white'})
plt.show()
```

### Piechart, Count of the different 'Sub-Category'

```python
plt.figure(figsize=(10,10))
plt.pie(useful_df['Sub-Category'].value_counts(), labels = useful_df['Sub-Category'].drop_duplicates(), 
        startangle = 0, autopct='%.1f%%', wedgeprops={'edgecolor':'white'})
plt.show()
```
